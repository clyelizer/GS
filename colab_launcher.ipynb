```python
# 2. Configurer le tunnel et lancer l'application
import os
from pyngrok import ngrok
from gestion_scolaire.app import create_app
from gestion_scolaire.database import init_db
import threading

# Authentification ngrok (Optionnel mais recommand√©)
# Decommentez la ligne suivante et ajoutez votre token si vous en avez un
# ngrok.set_auth_token("VOTRE_TOKEN_ICI")

# Terminer les tunnels existants
ngrok.kill()

# Cr√©er le tunnel
public_url = ngrok.connect(5000).public_url
print(f"üöÄ Application accessible ici : {public_url}")

# Configuration de l'application
os.environ['FLASK_ENV'] = 'development'
app = create_app()

# Initialiser la DB si besoin
if not os.path.exists('school.db'):
    with app.app_context():
        init_db(app)

# Lancer le serveur
app.run(port=5000)
```
